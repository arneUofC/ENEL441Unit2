# Loop-Shaping - back to feedback equation. 

So far have looked at increasing system performance, but what about the other advantages of control? (disturbance rejection and robustness)?

# Balancing Priorities

Recall $\varepsilon$ is the difference between desired output ($r$) and actual output ($y$):

$$
\begin{align*}
\mathcal{E}(s) &=R(s) - Y(s) \\
&=S(s)r(s) + T(s)n(s) - S(s)P(s)d(s)
\end{align*}
$$

Ideally, we would like both $S$ and $T$ to be small so that we have good reference tracking, disturbance rejection and noise rejection. 

Recall: this is not possible since $S(s) + T(s) = 1$, so a reduction in $S$ necessarily means an increase in $T$. 

# Key Insignt 1:  
take a frequency domain point of view!

$$
\begin{align*}
S(j \omega) + T(j \omega) = 1
\end{align*}
$$

This equation still must hold, but importantly it must hold at every frequency.

# Key Insight 2:

For many applications:
- references are low frequency signals,
- disturbances are relatively low frequency,
- measurement noise (sensor noise) is high frequency.

# Examples of References, Sensor Noise and Disturbances in Time and Frequency Domain

# Combine insights:

- For low frequencies, want good reference tracking and good disturbance rejection
- For high frequencies, want good sensor noise rejection

In otherwords:
- For low frequencies, want $S$ to be small
- For high frequencies, want $T$ to be small

The trade off:
- For low frequencies, want $S$ to be small (so $T$ will be large, but sensor noise has small power in this frequency range)
- For high frequencies, want $T$ to be small (so $S$ will be large, but references and disturbances have small power in this frequency range)

# How does $K$ affect frequency response of $S$ and $T$?

$$
\begin{align}
S(s) &= \frac{1}{1+L(s)}  = \frac{1}{1+P(s)K(s)} \\
T(s) &= \frac{L(s)}{1+L(s)} = \frac{P(s)K(s)}{1+P(s)K(s)}  
\end{align}
$$

# Example 1 : given a plant transfer function

Suppose that we have a model $P(s)$ for a plant. Then we can write:

$$
\begin{align*}
L(s) &= P(s)K(s) \\
\frac{k_{\ell}}{s} &= P(s)K(s) \\
K(s) &= \frac{k_{\ell}}{s P(s)}
\end{align*}
$$


Caution! We are inverting the plant transfer function. This means that $P(s)$ cannot have any non-minimum phase zeros! Why?

```python
import control as ct
import enel441_utilities as eu
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

# set system parameters
J1 = 2
J2 = 5
k = 1
b = 0.5

# transfer function numerator and denominator (from equations)
P_num = np.array([1])
P_den = np.array([J1*J2/k, (J1+J2)*b/k, (J1+J2+b**2/k), 2*b, 0])

k_ell = 1
K_num = P_den*k_ell
K_den = np.array([1, 0])*P_num

K = ct.tf(K_num, K_den)  
P = ct.tf(P_num, P_den)

S = 1/(1+P*K)
T = P*K/(1 + P*K)

N = 10000
t = np.linspace(0,100,N)
rr = np.random.randn(N)
b, a = sp.signal.butter(5, 0.01, 'low')
r = sp.signal.lfilter(b, a, rr)

b, a = sp.signal.butter(5, 0.05, 'low')
d = sp.signal.lfilter(b, a, np.random.randn(N))
d += np.linspace(0,1,N)
d *= 0.1  

n = 0.05*np.random.randn(N)

# Simulate closed-loop response
t, y_cl_r = ct.forced_response(T,t,r) 
t, y_cl_n = ct.forced_response(T,t,n) 
t, y_cl_d = ct.forced_response(P*S,t,d) 

y_cl = y_cl_r + y_cl_n
 
# plot results
fig,ax = plt.subplots(1,3, figsize=(15,5) )
ax[0].plot(t,r)
ax[0].set_xlabel('Time (s)')
ax[0].set_title('Desired Output')
ax[1].plot(t,d)
ax[1].set_xlabel('Time (s)')
ax[1].set_ylabel('Disturbance')
ax[1].set_title('Disturbance')   
ax[2].plot(t,n)    
ax[2].set_xlabel('Time (s)')
ax[2].set_ylabel('Measurement Noise')
ax[2].set_title('Measurement Noise')
        
fig.tight_layout()
    
fig,ax = plt.subplots(1,1, figsize=(15,5))
ax.plot(t,r, label='Desired Output') 
ax.plot(t,y_cl, label='Measured Output Using Closed Loop Control') 
ax.set_xlabel('Time (s)')
#ax.set_ylabel('Speed (km/h)')
ax.set_title('Comparison of Desired and Measured Output')
plt.legend()
```

# Example 2: given a Bode plot of plant

Often it is possible to obtain a Bode Plot of a system from data (no need to model the system using equations). In this situation, we can use a graphical loop-shaping approach.

```python
import control as ct

P = ct.tf(1,[1, 2, 5])
mag,phase,w=ct.bode(P)
```

# Step 1 - Place poles and zeros so that the loop transfer function looks like a straight line with slope -20dB

Strategy: Visually try to place poles and zeros such that the Magnitude response of $P(s)K(s)$ looks like $\frac{1}{s}$.

Suppose we have selected $k_{\ell} = 1$ based on the typical reference signal bandwidth.

For the plot shown above we see that there are two complex conjugate poles with real part $\alpha \approx 2$. So place two zeros at $-2$. Then the controller becomes:

$$ K= \frac{(s+2)^2}{s} $$

The Bode plot of the resulting loop transfer function is:


```python
K = ct.tf([1, 4, 4], [1, 0])
mag,phase,w=ct.bode(P*K)
```

# Step 2 - Adjust the scale of the plot to achieve the required $k_l$

- $k_{\ell}$ adjusts the scale of the Loop transfer function $L(s)=\frac{k_{\ell}}{s}$. 
- Need to scale the Bode Plot of the Loop transfer so that we have the correct $k_{\ell}$ that we desire (selected based on reference signal and sensor noise).

One way to read the value of $k_{\ell}$ off the Bode Plot of the loop transfer function:
1. Suppose that the loop transfer function is (nearly) a straight line with slope -20dB.
2. This means that its transfer function is $\frac{k_{\ell}}{s}$
3. Consider the frequency as which the Loop transfer function has a magnitude of 1: 
$$
\begin{align*}
    |L(j\omega_{1})| &= \Big|\frac{k_{\ell}}{j\omega_1}\Big|
    1 &= \frac{k_{\ell}}{\omega_1} \\
    \omega_1 &= k_{\ell}
\end{align*}
$$
So we conclude that the frequency where the loop transfer function has magnitude of $1$, denoted as $\omega_1$ is equal to $k_{\ell}$. Thus we can read the value of $k_{\ell}$ off the Bode Plot. 

This means that we can scale the Bode Plot of the Loop Transfer function is such a way that the frequency at which that magnitude is equal to $1$ is equal to $k_{\ell}$.

This is denoted in the plot below.



```python
K = ct.tf([1, 4, 4], [1, 0])

mag,phase,w=ct.bode(P*K,plot=False)

ii = 0
for m1 in mag:
    ii += 1
    if m1 < 1:
        break
 
fig,ax=plt.subplots(2,1)
ax[0].loglog(w,mag)
ax[0].set_ylabel('Magnitude')
ax[0].grid(visible=True, which='both')
ax[0].set_title('Bode Plot of Loop Transfer Function')
ax[0].plot(w,np.ones(len(w)),'r--')
ax[0].plot([w[ii], w[ii]], [ax[0].get_ylim()[0], ax[0].get_ylim()[1]], 'r--')
ax[1].semilogx(w,phase*180/np.pi)
ax[1].set_ylabel('Phase (deg)')
ax[1].set_xlabel('Frequency (rad/s)')
ax[1].grid(visible=True, which='both')
```

And the Bode plot of the Sensitivity and Complementary Sensitivity Function are:

```python
mag,phase,w=ct.bode(1/(1 + P*K))
mag,phase,w=ct.bode(P*K/(1 + P*K))
```

Although these plots don't look quite as nice as the previous ones, they do have the main feature that $S(j \omega)$ is small for $0 - 1$ rad/s, and $T(j \omega)$ is small for $1-10$ rad/s.

# Special Transfer Functions In Addition to Sensitivity and Complementary Sensitivity Functions

![feedback loop](feedbackLoop.png "feedback loop")

Consider the standard closed loop with external variables $r(t)$, $d(t)$, and $n(t)$. Derive expressions for $Y(s)$, $U(s)$, and $\mathcal{E}(s)$. How many unique transfer functions appear in these expressions?

From the figure of a closed loop system, the following expressions can be derived
$$
\begin{align*}
Y(s) &= \frac{P(s)K(s)}{1 + P(s)K(s)}R(s) + \frac{P(s)}{1 + P(s)K(s)}D(s) - \frac{P(s)K(s)}{1 + P(s)K(s)}N(s), \\
U(s) &= \frac{K(s)}{1 + P(s)K(s)}R(s) + \frac{1}{1 + P(s)K(s)}D(s) - \frac{K(s)}{1 + P(s)K(s)}N(s),  \\
\mathcal{E}(s) &= \frac{1}{1 + P(s)K(s)}R(s) - \frac{P(s)}{1 + P(s)K(s)}D(s) + \frac{P(s)K(s)}{1 + P(s)K(s)}N(s), 
\end{align*}
$$
where there are only 4 transfer functions to consider (aka the Gang of 4):
$$
\begin{align*}
S(s) &= \frac{1}{1 + P(s)K(s)} \\
T(s) &= \frac{P(s)K(s)}{1 + P(s)K(s)} \\
S(s)P(s) &= \frac{P(s)}{1 + P(s)K(s)} \\
S(s)K(s) &= \frac{K(s)}{1 + P(s)K(s)} 
\end{align*}
$$

# What Do They Mean?

Main uses:
- $S(s)$ determines what portion of the reference signal contributes to the tracking error.
- $T(s)$ determines what portion of the sensor noise contributes to the tracking error.
- $P(s)S(s)$ determines what portion of the disturbance $D(s)$ affects the tracking error.
- $K(s)S(s)$ determines how the reference signal affects the input to the plant.

# Summary of important topics:
- Feedback equation
- Advantages of control:
    - Increase performance
    - Disturbance Rejection
    - Robustness to modelling errors/changes in the model
- Open loop vs. closed loop system and corresponding transfer function
- Sensitivity Function
    - determines tracking error
- Complementary Sensitivity Function
    - transfer function from reference to output
    - governs measurement noise rejection
- Proportional Control - Choose $K$ equal to a constant
    - Attempt to make $S$ small by increasing $K$
    - Can change step response characteristics
    - Problems with steady state errors (can be very large)
- Root Locus - map closed-loop poles that are possible for a proportional controller onto s-plane
- Loop-Shaping
    - Shape Sensitivity and Complementary Sensitivity Functions per frequency to obtain satisfactory reference tracking and measurement noise rejection
    - One attractive loop transfer function is $\frac{k_{\ell}}{s}$. Can design $K$ to achieve this loop transfer function.
